# 4 - Correlação
**Direto ao ponto**: correlação é o coeficiente de interdependência entre duas ou mais variáveis.

Em muitos casos, quando analisamos algumas variáveis, essa interdependência é muito clara. Exemplo:
- Massa corporal e altura,
- Consumo de combustível e distância percorrida,
- Número de anúncios e produtos vendidos,
- Tempo de estudo e desempenho na avaliação.

Entretanto, algumas vezes essa correlação pode não ser muito clara.

## Correlações positivas

```python
# Importa o numpy
import numpy as np

# Cria uma seed
np.random.seed(1)

# array de 1000 números inteiros entre 0 e 50
x = np.random.randint(0, 50, 1000)

# Correlação positiva com algum ruído
# Função de normalização dos dados
y = x + np.random.normal(0, 10, 1000)

# Retorna um array de correlação positiva de interdependência entre x e y
np.corrcoef(x, y)
```

```python
# import matplotlib
import matplotlib.pyplot as plt

# Cria um gráfico com a relação entre os dados de x e y
plt.scatter(x, y)
```

## Correlações negativas

```python
# Array de 1000 números inteiros entre 0 e 50
x = np.random.randint(0, 50, 1000)

# Correlação negativa com algum ruído
y = 100 - x + np.random.normal(0, 5, 1000)

# Retorna um array de correlação negativa de interdependência entre x e y
np.corrcoef(x, y)
```

```python
# Cria um gráfico com a relação entre os dados de x e y
plt.scatter(x, y)
```

## Correlação fraca ou sem correlação

```python
# array de 1000 números inteiros entre 0 e 50
x = np.random.randint(0, 50, 1000)
# array de 1000 números inteiros entre 0 e 50
y = np.random.randint(0, 50, 1000)

# Retorna um array de correlação entre x e y
np.corrcoef(x, y)
```

```python
# Cria um gráfico com a relação entre os dados de x e y
plt.scatter(x, y)
```

## Correlação é diferente de causalidade!
